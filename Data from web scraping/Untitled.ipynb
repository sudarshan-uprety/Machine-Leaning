```python
name=[]
rating=[]
reviews=[]
c_type=[]
hq=[]
old=[]
employees=[] 
for i in company:
    
    name.append(i.find('h2').text.strip())
    rating.append(i.find_all('p',class_='rating')[0].text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip())
    c_type.append(i.find_all('p','infoEntity')[0].text.strip())
    hq.append(i.find_all('p','infoEntity')[1].text.strip())
    old.append(i.find_all('p','infoEntity')[2].text.strip())
    employees.append(i.find_all('p','infoEntity')[3].text.strip())
    
d={'name':name,'rating':rating,'reviews':reviews,'ctype':c_type,'hq':hq,'old':old,'employees':employees}
df=pd.DataFrame(d)
df
```

```python
final=pd.DataFrame()
for j in range(333,334):
    url='https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(url,headers=headers).text
    
    soup=BeautifulSoup(webpage,'lxml')
    
    company=soup.find_all('div',class_='company-content-wrapper')
    
    name=[]
    rating=[]
    reviews=[]
    c_type=[]
    hq=[]
    old=[]
    employees=[] 
    for i in company:
        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)
            
            
        try:
            rating.append(i.find_all('p',class_='rating')[0].text.strip())
        except:
            rating.append(np.nan)
            
            
        try:
            reviews.append(i.find('a',class_='review-count').text.strip())
        except:
            reviews.append(np.nan)
            
            
        try:
            c_type.append(i.find_all('p','infoEntity')[0].text.strip())
        except:
            c_type.append(np.nan)
            
            
        try:
            hq.append(i.find_all('p','infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)
        
        
        try:
            old.append(i.find_all('p','infoEntity')[2].text.strip())
        except:
            old.append(np.nan)
            
            
        try:
            employees.append(i.find_all('p','infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)

    d={'name':name,'rating':rating,'reviews':reviews,'ctype':c_type,'hq':hq,'old':old,'employees':employees}
    df=pd.DataFrame(d)
    final=final.append(df)
```

```python
final
```

```python
i_data=pd.read_csv('final_company_data.csv')
```

```python
test_final=pd.concat([i_data,final],axis=0)
```
