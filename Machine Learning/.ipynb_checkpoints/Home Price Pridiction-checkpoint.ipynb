```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
```

```python
df=pd.read_csv('home_data.csv')
df
```

```python
model_parameters={
    'MultinomialNB':{
        'model':MultinomialNB(),
        'parameters':{}
    }
     #   'RandomForestClassifier':{
    #    'model':RandomForestClassifier(),
   #     'parameters':{
  #          'n_estimators':[1,5,10,15,20]
 #       }

#},
    #'GaussianNB':{
     #   'model':GaussianNB(),
      #  'parameters':{
       #     'var_smoothing': np.logspace(0,-9, num=100)
        #}
    #}
}
```

```python
scores=[]

for model_name,mp in model_parameters.items():
    clf=GridSearchCV(mp['model'],mp['parameters'],cv=5,return_train_score=False)
    clf.fit(data,df.price)
    scores.append({
        'Model Name':model_name,
        'Best Score':clf.best_score_,
        'Best Parameters':clf.best_params_
    })
df=pd.DataFrame(scores,columns=['Model Name','Best Score','Best Parameters'])
df
```
