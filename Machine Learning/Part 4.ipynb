# Hyper parameter Tuning (GridSearchCV)

```python
from sklearn import svm,datasets
iris=datasets.load_iris()
```

```python
import pandas as pd
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['flower']=iris.target
df['flower']=df['flower'].apply(lambda x: iris.target_names[x])
df[47:52]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.3)
```

```python
model=svm.SVC(kernel='rbf',C=30,gamma='auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='linear',C=10,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=10,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=20,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
import numpy as np
kernals=['rbf','linear']
C=[1,10,20]
avg_score={}

for kval in kernals:
    for cval in C:
        cv_score=cross_val_score(svm.SVC(kernel=kval,C=cval,gamma='auto'),iris.data,iris.target,cv=5)
        avg_score[kval+'_'+str(cval)]=np.average(cv_score)
        
avg_score
```

```python
from sklearn.model_selection import GridSearchCV
c_cal=[]
for x in range(1,100):
    c_cal.append(x)
clf=GridSearchCV(svm.SVC(gamma='auto'),{
    'C':c_cal,
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)

clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df=pd.DataFrame(clf.cv_results_)
```

```python
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs=RandomizedSearchCV(svm.SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False,n_iter=2
                     )
rs.fit(iris.data,iris.target)
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_parameters={
    'svm':{
        'model':svm.SVC(gamma='auto'),
        'params':{
            'C':[1,10,20],
            'kernel':['rbf','linear']
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    }
}
```

# Hyper parameter Tuning (GridSearchCV) EXERCISE

```python
import pandas as pd
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
model_parameters={
    'svm':{
        'model':svm.SVC(gamma='auto'),
        'parameters':{
            'C':[1,5,10,15,20],
            'kernel':['rbf','linear']
        }
    }
    ,
    'RandomForestClassifier':{
        'model':RandomForestClassifier(),
        'parameters':{
            'n_estimators':[1,5,10,15,20]
        }
    },
    'logestic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'parameters':{
            'C':[1,5,10,15,20]
        }
    },
    'GaussianNB':{
        'model':GaussianNB(),
        'parameters':{
            'var_smoothing': np.logspace(0,-9, num=100)
        }
    },
    'MultinomialNB':{
        'model':MultinomialNB(),
        'parameters':{}
    },
    'DecissionTree':{
        'model':DecisionTreeClassifier(),
        'parameters':{
            'criterion': ['gini','entropy']
        }
        
    }
}
```

```python
scores=[]

for model_name,mp in model_parameters.items():
    clf=GridSearchCV(mp['model'],mp['parameters'],cv=5,return_train_score=False)
    clf.fit(digits.data,digits.target)
    scores.append({
        'Model Name':model_name,
        'Best Score':clf.best_score_,
        'Best Parameters':clf.best_params_
    })
df=pd.DataFrame(scores,columns=['Model Name','Best Score','Best Parameters'])
df
```

# L1 and L2 Regularization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('Melbourne_housing_FULL.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
cols_to_use=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
dataset=dataset[cols_to_use]
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zer0=['Propertycount','Distance','Bedroom2','Bathroom','Car']
dataset[cols_to_fill_zer0]=dataset[cols_to_fill_zer0].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize']=dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea']=dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset.head()
```

```python
x=dataset.drop('Price',axis='columns')
y=dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(train_x,train_y)
```

```python
reg.score(test_x,test_y)
```

```python
reg.score(train_x,train_y)
```

```python
from sklearn import linear_model
lego_reg=linear_model.Lasso(alpha=50,max_iter=100,tol=0.1) #for l1 regularization
lego_reg.fit(train_x,train_y)
```

```python
lego_reg.score(test_x,test_y)
```

```python
lego_reg.score(train_x,train_y)
```

```python
from sklearn.linear_model import Ridge #for l2 regularization

redg_reg=Ridge()
redg_reg.fit(train_x,train_y)
```

```python
redg_reg.score(test_x,test_y)
```

```python
redg_reg.score(train_x,train_y)
```
